```python
import sqlite3
#Connecting to sqlite
conn = sqlite3.connect('Authorize.db')

#Creating a cursor object using the cursor() method
cursor = conn.cursor()

#Doping EMPLOYEE table if already exists.
cursor.execute("DROP TABLE IF EXISTS EMPLOYEE")
cursor.execute("DROP TABLE IF EXISTS EMPLOYEE1")
cursor.execute("DROP TABLE IF EXISTS EMPLOYEE2")
cursor.execute("DROP TABLE IF EXISTS EMPLOYEE3")
#Creating table as per requirement
sql ='''CREATE TABLE EMPLOYEE(
   ID INT PRIMARYKEY NOT NULL,
   FIRST_NAME CHAR(20) NOT NULL,
   EMAIL CHAR(20),
   AGE INT,
   SEX CHAR(1),
   INCOME FLOAT
)'''
cursor.execute(sql)
print("Table created successfully........")

#Commit your changes in the database
conn.commit()

#Closing the connection
conn.close()
```

```python
import numpy as np
from pyzbar.pyzbar import decode
import cv2
#import bol
# most imp for decoding and encoding

#img=cv2.imread("res/qrcode/barauth.jpg")
# code=decode(img)
# web cam
cap = cv2.VideoCapture(0)
cap.set(3, 640)
cap.set(4, 480)

mydatalist=resultk
print(mydatalist)

while(cap.isOpened()):#Returns true when the path is existing else False
    # ret returns true if frame is avl else false
    ret,frame = cap.read()
    if ret==True:
          
        #print(cv2.CAP_PROP_FRAME_WIDTH)
        #print(cv2.CAP_PROP_FRAME_HEIGHT)
        #gray= cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        #cv2.imshow('video',frame)
        #print(gray)
        for barcode in decode(frame):
            print(barcode.data)
            print(barcode.rect)
            myData = barcode.data.decode('utf-8')  # convert it to string
                # draw bounding box
            pts = np.array([barcode.polygon], np.int32)
            pts = pts.reshape(-1, 1, 2)

                # check Authorisation
            if myData in mydatalist:
                Myoutput='Authorised'
                Mycolor=(0,255,0)
            else:
                Myoutput='Unauthorised'
                Mycolor = (0, 0, 255)
            cv2.polylines(frame, [pts], True, (255, 0, 255), 5)
            pts2 = barcode.rect
            cv2.putText(frame, Myoutput, (pts2[0], pts2[1]), cv2.FONT_HERSHEY_PLAIN, 0.9,Mycolor, 2)

        cv2.imshow("Results", frame)
            #cv2.waitKey(0)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()

```
